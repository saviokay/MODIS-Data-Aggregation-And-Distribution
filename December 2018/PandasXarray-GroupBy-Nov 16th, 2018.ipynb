## Pandas : Implemenatation

```python
mod03_name = '/Users/saviosebastian/Documents/Project/CMAC/MYD03.A2002185.0000.061.2017362174430.hdf'
mod06_name = '/Users/saviosebastian/Documents/Project/CMAC/MYD06_L2.A2002185.0000.061.2018003215042.hdf'
print(mod03_name)
print(mod06_name)
```

```python
rootgrp = Dataset(mod03_name, "r", format="NETCDF3")
latitude = rootgrp.variables["Latitude"][:,:] 
longitude = rootgrp.variables["Longitude"][:,:]

rootgrp1 = Dataset(mod06_name, "r", format="NETCDF3")
cloud_mask_allbytes = rootgrp1.variables["Cloud_Mask_1km"][:,:,:] 
```

```python
londf=pd.DataFrame(np.array(longitude).byteswap().newbyteorder())
latdf=pd.DataFrame(np.array(latitude).byteswap().newbyteorder())

cat = np.asarray(cloud_mask_allbytes)
samcat = cat[:][:][:]
samcatran = samcat.transpose(2,0,1).reshape(3,-1)
samcatran2 = samcat.transpose(2,0,1)
s3 = samcatran2[:][:][0]
df34 = pd.DataFrame(s3)
catdf=pd.DataFrame(np.array(df34).byteswap().newbyteorder())
```

```python
londfstack = londf.stack()
latdfstack = latdf.stack()
catdfstack = catdf.stack()

maindf = pd.DataFrame()
```

```python
maindf['Latitude'] = latdfstack
maindf['Longitude'] = londfstack
maindf['CloudFraction'] = catdfstack
maindf['Latitude-Int'] = maindf.Latitude.astype(np.int16)
maindf['Longitude-Int'] = maindf.Longitude.astype(np.int16)
```
